## Import Modules

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Request page source from URL

```python
url = "https://www.imdb.com/chart/top/"
```

```python
r=requests.get(url)
```

```python
r
```

```python
r.status_code
```

### display the page source code

```python

r.content
```

```python
print(r.text)
```

```python
soup=BeautifulSoup(r.content,"lxml")
```

```python
soup
```

```python
print(soup.prettify())   
```

### scrap movie names

```python

scraped_movies = soup.find_all('td', class_='titleColumn')
scraped_movies
```

### parse movie names

```python

movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
len(movies)
```

### scrap rating for movies

```python

scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings
```

### parse ratings

```python

ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings.append(rating)
ratings
```

```python
len(ratings)
```

## Store the Scraped Data

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] = ratings
data.head()
```

```python
data.to_csv('IMDB Top Movies.csv', index=False)
```
